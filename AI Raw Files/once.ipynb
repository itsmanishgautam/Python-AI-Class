```python
import urllib.request
from bs4 import BeautifulSoup

job_titles = []
companies = []
locations = []
categories = []

# User-Agent string to avoid being blocked
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'

for page_num in range(1, 14):
    url = f'https://www.jobsnepal.com/jobs?page={page_num}'

    request = urllib.request.Request(url, headers={'User-Agent': user_agent})
    response = urllib.request.urlopen(request)
    html_content = response.read().decode('utf-8')

    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Extract job titles, company names, and other information
    titles = soup.find_all('h2', class_='job-title')
    companies_divs = soup.find_all('div', class_='card-inner')
    tags = soup.find_all('div', class_=['icon-price-tags mr-2 text-success icon-sm', 'd-flex align-items-center pl-1 pr-1 py-1'])

    job_titles.extend([title.text.strip() for title in titles])

    # Extracting company names from card-inner divs
    for div in companies_divs:
        company = div.find('a').get('title') if div.find('a') else 'N/A'
        companies.append(company)

    # Extracting additional information
    doubledata = [tag.find('div').text.strip() for tag in tags]
    locations.extend(doubledata[::2])
    categories.extend(doubledata[1::2])

# Print all the scraped data
for job_title, company, location, category in zip(job_titles, companies, locations, categories):
    print("Job Title:", job_title)
    print("Company:", company)
    print("Location:", location)
    print("Category:", category)
    print()
```
