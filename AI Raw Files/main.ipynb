```python
import csv
import nltk
from nltk.tokenize import word_tokenize

def parse_csv_and_tokenize(file_path):
    with open(file_path, newline='') as csvfile:
        reader = csv.reader(csvfile)
        header = next(reader)
        data_list = []
        for row in reader:
            row_tokens = [word_tokenize(cell.strip()) for cell in row]
            row_dict = {header[i]: row_tokens[i] for i in range(len(header))}
            data_list.append(row_dict)
    return data_list

# Parse and tokenize each CSV data
nepal_data_tokenized = parse_csv_and_tokenize('csv/nepal_data.csv')
gold_data_tokenized = parse_csv_and_tokenize('csv/Gold_Silver_Cost.csv')
bank_interest_tokenized = parse_csv_and_tokenize('csv/bank_interest.csv')
nepse_data_tokenized = parse_csv_and_tokenize('csv/nepse_data.csv')

# Output the tokenized lists of dictionaries
print("Nepal Data:")
for item in nepal_data_tokenized:
    print(item)

print("\n  Bank Interest Data:")
for item in bank_interest_tokenized:
    print(item)
```
